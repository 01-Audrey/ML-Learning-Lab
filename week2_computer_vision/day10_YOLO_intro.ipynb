```python
# ============================================================
# LOADING YOLO MODEL
# ============================================================

from ultralytics import YOLO
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

print("\n" + "=" * 60)
print("LOADING YOLOv8 MODEL")
print("=" * 60)

print("""
Loading pre-trained YOLOv8 model...
This will download the model weights (~6 MB) on first run.
""")

# Load YOLOv8 nano model (smallest, fastest)
model = YOLO('yolov8n.pt')  # 'n' = nano (smallest)

print("\nâœ… YOLOv8 loaded successfully!")
print(f"Model: YOLOv8-nano")
print(f"Pre-trained on: COCO dataset (80 classes)")
print(f"Ready to detect: people, cars, animals, and more!")

# Show what classes YOLO can detect
print("\n" + "=" * 60)
print("YOLO CAN DETECT 80 CLASSES:")
print("=" * 60)

classes = model.names  # Dictionary of class names
print("\nSample classes:")
for i in [0, 1, 2, 3, 16, 17, 24, 25, 26, 27]:
    print(f"  {i}: {classes[i]}")
print(f"  ... and {len(classes) - 10} more!")

print("""
NOTE: YOLO is pre-trained on general objects.
For Project 1 (helmets, vests), we'll fine-tune it on our data!

Let's test it on a sample image! ğŸ¯
""")
```

```python
# ============================================================
# RUNNING DETECTION AND VISUALIZATION
# ============================================================

print("\n" + "=" * 60)
print("RUNNING YOLO DETECTION")
print("=" * 60)

# Run inference on sample image
results = model('https://ultralytics.com/images/bus.jpg')

print("âœ… Detection complete!")

# Get the results
result = results[0]  # First (only) image
num_detections = len(result.boxes)

print(f"\nFound {num_detections} objects!")

# Show each detection
classes = model.names
for i, box in enumerate(result.boxes):
    cls = int(box.cls[0])
    conf = float(box.conf[0])
    class_name = classes[cls]
    
    print(f"\n{i+1}. {class_name}")
    print(f"   Confidence: {conf:.2f}")

# Visualize
print("\n" + "=" * 60)
print("VISUALIZING DETECTIONS")
print("=" * 60)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Original image
from PIL import Image
import requests
from io import BytesIO

response = requests.get('https://ultralytics.com/images/bus.jpg')
img_original = Image.open(BytesIO(response.content))

axes[0].imshow(img_original)
axes[0].set_title('Original Image', fontsize=14, fontweight='bold')
axes[0].axis('off')

# Image with detections (YOLO draws boxes automatically!)
img_with_boxes = result.plot()
axes[1].imshow(img_with_boxes)
axes[1].set_title(f'YOLO Detection ({num_detections} objects found)', 
                 fontsize=14, fontweight='bold')
axes[1].axis('off')

plt.tight_layout()
plt.show()

print("""
ğŸ‰ YOU JUST RAN YOUR FIRST OBJECT DETECTION!

WHAT YOLO DID:
âœ… Looked at the image ONCE (~200ms)
âœ… Detected ALL objects (people, bus, etc.)
âœ… Drew bounding boxes with labels
âœ… Gave confidence scores

THIS IS THE SAME TECHNOLOGY YOU'LL USE FOR PROJECT 1!

Instead of detecting "person" and "bus", you'll detect:
  - "helmet"
  - "vest"
  - "person"

Same YOLO, different training data! ğŸ—ï¸ğŸ¯
""")
```
